```python
import requests
import json
import pandas as pd 

#Webiste where I was able to find the API and documentation necessary to acess the data 
#https://docs.coincap.io/

#THIS API DOES NOT NEED A API KEY 

#This is the URL which I would like it to access. 
url = "http://api.coincap.io/v2/assets"

payload={}
headers = {}

#No key necessary
response = requests.request("GET", url, headers=headers, data=payload)

data=response.json()
#print(data)

formatted_json=json.dumps(data,sort_keys=True, indent=5)
print(formatted_json)
```

```python
print(f'Status Code: {response.status_code}')
```

```python
#Printing all the different symbols withnin the data. 
for x in data['data']:
  print(x['symbol'])
```

```python
#This is to show all the different, cryptocurrencies. 

Crypto=[]
Symbol=[]
PercentChange=[]
MarketCap=[]
PriceUSD=[]
Rank=[]
Supply=[]
Volume=[]

for x in data["data"]:
  Crypto.append(x['id'])
  Symbol.append(x['symbol'])
  PercentChange.append(float(x['changePercent24Hr']))
  MarketCap.append(float(x['marketCapUsd']))
  PriceUSD.append(float(x['priceUsd']))
  Rank.append(float(x['rank']))
  Supply.append(float(x['supply']))
  Volume.append(float(x['volumeUsd24Hr']))
```

```python
df = pd.DataFrame(list(zip(Crypto, Symbol,PercentChange,MarketCap,PriceUSD,Rank,Supply,Volume)),
               columns =['Cryptocurrency','Stock Symbol', '24 Hour Percent Change','MarketCap','Price $USD','Rank','Supply','Volume'])

df.set_index('Rank')
```
