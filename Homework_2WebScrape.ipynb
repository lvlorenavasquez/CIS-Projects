```python
# Importing the libraries needed to webscrape
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
#Webpage that I will be Webscraping
URL = 'https://www.worldometers.info/world-population/'
#Requesting the program to get the url 
page = requests.get(URL)
#BeautifulSoup will help in being able to read the data
soup = BeautifulSoup(page.content, 'html.parser')
page.content
#Printing the pageconent in a different manner, making it easier to read. 
print(soup.prettify())
```

```python
#Find the table within the page
table = soup.find('table')
print(table)
```

```python
#printing the table rows and find the information
print(table.findAll('tr'))
```

```python
#Creating empty lists for the columns 
Year=[]
Population=[]
YearlyPercentChange = []
YearlyChange=[]
MedianAge=[]
FertilityRate=[]
Density=[]
UrbanPopPercent=[]
UrbanPop=[]

#Finding all the table rows 
rows = table.findAll('tr')
#iterating through the rows and finding the table data 
for row in rows:
 cells = row.findAll('td')
 #I know there are 9 columns, so there are 9 cells to iterate
 if(len(cells) == 9):
  Year.append(float(cells[0].find(text=True)))
  Population.append(cells[1].find(text=True))
  YearlyPercentChange.append(cells[2].find(text=True))
  YearlyChange.append(cells[3].find(text=True))
  MedianAge.append(float(cells[4].find(text=True)))
  FertilityRate.append(float(cells[5].find(text=True)))
  Density.append(float(cells[6].find(text=True)))
  UrbanPopPercent.append(cells[7].find(text=True))
  UrbanPop.append(cells[8].find(text=True))



```

```python
#Creating a new list to drop the % in the td tags 
NewYearlyPercChange=[]
for new in YearlyPercentChange:
  NewYearlyPercChange.append(float(new.split()[0]))
```

```python
#Ziping the lists together to greate a dataframe 
df = pd.DataFrame(list(zip(Year,Population,YearlyPercentChange,YearlyChange,MedianAge,FertilityRate,Density,UrbanPopPercent,UrbanPop
)),
               columns =['Year','Population','Yearly%Change','Yearlychange','MedianAge','FertilityRate','Density','UrbanPopulation%','Urban Population'])
```

```python
df.set_index('Year')
#Temporarily setting the index to year
```
