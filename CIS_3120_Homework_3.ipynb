```python
import pandas as pd 
import json 
import requests 
```

```python
#set the url/endpoint to which we are making a request.
URL = 'https://api.covid19api.com/summary'
#make a request to the API using the url 
response = requests.get(URL)
#status code of 200 means the request was successful
print(f'Status Code{response.status_code}')
#by default, the api returns the data in json format
data = response.json()
print(data)
#json.dumps() 
formatted_json = json.dumps(data, sort_keys=True, indent=5)
print(formatted_json)
```

```python
#My empty list 
Country = []
CountryCode = []
Date = []
ID =[]
NewConfirmed =[]     
NewDeaths = []
NewRecovered = []
TotalConfirmed = []
TotalDeaths = []
TotalRecovered = []

#Making a loop to append my list 
info = data['Countries'][0:]
for i in info:
    Country.append(i['Country'])
    CountryCode.append(i['CountryCode'])
    Date.append(i['Date'])
    ID.append(i['ID'])
    NewConfirmed.append(i['NewConfirmed'])
    NewDeaths.append(i['NewDeaths'])
    NewRecovered.append(i['NewRecovered'])
    TotalConfirmed.append(i['TotalConfirmed'])
```

```python
#Creating my dictionary 
Dictionary_Columns = {'Country': Country, 'CountryCode' : CountryCode, 'Date' : Date, 'ID':ID, 'NewConfirmed': NewConfirmed, 'NewDeaths' : NewDeaths, 'NewRecovered' : NewRecovered, 'TotalConfirmed' : TotalConfirmed  }
#creating my dataframe
Countries_df=pd.DataFrame(Dictionary_Columns)
```

```python

from bs4 import BeautifulSoup

#The Url of population by country 
URL = 'https://www.worldometers.info/world-population/population-by-country/'
# get web data
page = requests.get(URL)
soup = BeautifulSoup(page.content, 'html.parser')
#Protomg tje website for it to be readable 
print(soup.prettify())
```

```python
results = soup.find(id='example2')
```

```python
#creating empty list
headings = []
#extract headings
th_headings = results.find_all("th")
#loop for my header
for th_heading in th_headings:
    th_heading = (th_heading.text).strip("\n")
    headings.append(th_heading)
print(headings)
```

```python
#empty list for my data
Population = []
#to find my rows in the tag
for row in results.find_all('tr'):
    rows = []
    #for my cell in the td tag 
    for cell in row.find_all('td'):
        rows.append((cell.text).strip("\n"))
    #doing a loop for my headers and rows
    if(len(rows)> 0):
        data = {
                headings[0]: rows[0],
                headings[1]: rows[1],
                headings[2]: rows[2],
                headings[3]: rows[3],
                headings[4]: rows[4],
                headings[5]: rows[5],
                headings[6]: rows[6],
                headings[7]: rows[7],
                headings[8]: rows[8],
                headings[9]: rows[9],
                headings[10]: rows[10],            
        }

        Population.append(data)
```

```python
df = pd.DataFrame(Population)
df.head()
```

```python
#data frame changing the column name to make it easier to merge 
df.rename(columns={'Country (or dependency)': 'Country'}, inplace=True)
```

```python
#Merging the two dataframes 
Merged_df=Countries_df.merge(df,on='Country')
```
