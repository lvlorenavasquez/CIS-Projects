```python
#Pandas will help in allowing us to read the CSV file into a dataframe 
import pandas as pd

#matplotlib is a python library for basic graph plotting
import matplotlib.pyplot as plt

#seaborn is another python graphing library that is built upon matplotlib and allows you to optimize matplotlib's output
import seaborn as sns 

#switching from matplotlib's default graphing style to that of seaborn's. Seaborn displays graph lines, matplotlib doesn't
sns.set(color_codes = True)

#removing warnings from the output
import warnings
warnings.filterwarnings("ignore")
```

```python
#fish will be the dataframe we will use 
#just reading the csv file into the fish dataframe 
fish=pd.read_csv('Fish.csv')
```

```python
#checking the data by seeing the first 5 lines
fish.head()
```

```python
#check the shape of the dataset 
fish.shape
```

```python
#Correlation between Height and Width 
sns.set()
sns.scatterplot(data=fish, x='Height', y='Width')
```

```python
#Correlation between Height and Weight 

sns.scatterplot(data=fish, x='Weight', y='Height')
```

```python
#Correlation between Weight and Length3 
sns.scatterplot(data=fish, x='Weight', y='Length3')
```

```python
#A joint plot between Height and Width 
sns.jointplot(data=fish, x='Height', y='Width', size =5)
```

```python
#Correlation between Height and Width
#According to their species
sns.FacetGrid(fish, hue='Species', palette='husl', size=6).map(plt.scatter,'Height','Weight').add_legend()
```

```python
#Distribution of Height by the species 
sns.boxplot(x='Species',y='Height', palette='husl', data=fish)
```

```python
#Distribution of the ranges in Weight by species 
sns.boxplot(x='Species',y='Weight', palette='husl', data=fish)
```

```python
#Distribution of the lenght3 by species 
sns.boxplot(x='Species',y='Length3', palette='husl', data=fish)
```

```python
#Violin plot to show the distribution in Height by the species 
fig = plt.gcf()
fig.set_size_inches(14,8)
sns.violinplot(x='Species', y='Height', hue='Species',palette='husl', data=fish)
```

```python
#Violin plot to show the distribution in Height by the species 
#Including the swarm plot to show the data 
fig = plt.gcf()
fig.set_size_inches(15,8)
sns.violinplot(x='Species',y='Height',palette='husl',data=fish)
sns.swarmplot(x='Species',y='Height',data=fish,color='w',alpha=.9)
```

```python
#Shows the density of the variable height among the species 
sns.FacetGrid(fish, hue='Species', palette='husl', size=6).map(sns.kdeplot, 'Height').add_legend()
```

```python
#The pairplot of the variables against the species 
sns.pairplot(fish, hue='Species', palette='husl', size=3, diag_kind='kde')
```

```python
#Boxplot on the data for each variable by its species 
fish.boxplot(by='Species', figsize=(12, 10))
```

```python
from pandas.plotting import parallel_coordinates

#Because this relies heavily on the columns, 
#to be numerical values a new df needed to be 
#created just to include the numerical values. 

fish2=fish[['Species','Width','Height','Weight','Length3']]
parallel_coordinates(fish2,'Species')
```
